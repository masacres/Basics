```python
import torch
import torch.nn as nn
import seaborn as sns
```

```python
x = torch.arange(-5, 5, 0.1)
```

```python
# for binary classification
sns.lineplot(x=x, y=nn.Sigmoid()(x))
```

```python
sns.lineplot(x=x, y=nn.Tanh()(x))
```

```python
# for hidden layers
sns.lineplot(x=x, y=nn.ReLU()(x))
```

```python
# for hidden layers if there are dying ReLU problems
sns.lineplot(x=x, y=nn.LeakyReLU()(x))
```

```python
# for multiclass classification
sns.lineplot(x=x, y=nn.Softmax(dim=0)(x))
```
